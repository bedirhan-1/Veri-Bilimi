```python
import seaborn as sns

flights = sns.load_dataset("flights")

df = flights.copy()
```

```python
df.head()
```

```python
df
```

```python
df.describe().T
```

```python
df = df.pivot("month", "year", "passengers");
```

```python
df
```

```python
sns.heatmap(df);
```

```python
sns.heatmap(df, annot = True, fmt = "d");
```

```python
sns.heatmap(df, annot = True, fmt = "d", linewidths = 0.5);
```
