# Accessing to element with Fancy Index

```python
import numpy as np

v = np.arange(0, 30, 3)

v
```

```python
v[1]
```

```python
v[3]
```

```python
[v[1], v[3]]
```

```python
[v[1], v[3], v[5]]
```

```python
get = [1, 3, 5]
```

```python
v[get]
```

### Two dimension fancy

```python
m = np.arange(9).reshape(3, 3)
```

```python
m
```

```python
line = np.array([0, 1])
```

```python
column = np.array([1, 2])
```

```python
m[line, column]
```

```python
n = np.arange(15).reshape(3,5)
```

```python
n
```

```python
n[2, [3, 4]] # third line,  fourth and fifth element
```
