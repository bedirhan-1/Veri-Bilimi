# Processing subsets

```python
import numpy as np
a = np.random.randint(10, size = (5,5))
a
```

```python
subset_a = a[0:3, 0:2]
subset_a
```

```python
subset_a[0, 0] = 999
subset_a[1, 1] = 888
```

```python
subset_a
```

```python
a
```

### If you do not want to make changes to the original array

```python
m = np.random.randint(10, size = (5, 5))
m
```

```python
subset_b = m[0:3, 0:2].copy()
subset_b
```

```python
subset_b[0,0] = 999
```

```python
subset_b
```

```python
m
```
