# Splitting

```python
import numpy as np
```

```python
x = np.array([1, 2, 3, 42, 34, 67, 99, 12])
```

```python
np.split(x, [3, 5])
```

```python
a, b, c = np.split(x, [3, 5])
```

```python
a
```

```python
b
```

```python
c
```

```python
m = np.arange(16).reshape(4, 4)
```

```python
m
```

```python
np.vsplit(m, [2])
```

```python
top, bottom = np.vsplit(m, [2])
```

```python
top
```

```python
bottom
```

```python
np.hsplit(m, [2])
```

```python
right, left = np.hsplit(m, [2])
```

```python
right
```

```python
left
```
