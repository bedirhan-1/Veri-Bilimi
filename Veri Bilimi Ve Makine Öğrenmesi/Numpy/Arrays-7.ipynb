# Access to element with slicing

```python
import numpy as np
```

```python
a = np.arange(20,30)
```

```python
a[0:3]
```

```python
a[:3]
```

```python
a[3:]
```

```python
a[1::2]
```

```python
a[2::3]
```

# Multidimensions slice process
### Usage : 
#### arrayName[ row:row_end, column:column_end ]

```python
m = np.random.randint(10, size = (5,5))
```

```python
m
```

```python
m[:,0] # select all rows and print zero column
```

```python
m[:,1]
```

```python
m[:,3]
```

```python
m[0,:]
```

```python
m[0]
```

```python
m[1,:]
```

```python
m[0:2, 0:3]
```
