# İş Uygulaması:
#### Ürün satış olasılıklarının hesaplanması

#### Problem :
Bir yatırım-toplantı öncesinde gelecek ay ile ilgili satışların belirli değerlerde gerçekleşmesi olasılıkları hesaplanması isteniyor.

#### Detaylar :
* Dağılım 'Normal Dağılım'.
* Aylık ortalama satış sayısı 80K, standart sapması 5K
* 90K'dan fazla satış yapma olasılığı nedir?

```python
from scipy.stats import norm
```

```python
# 90'dan fazla olma olasılığı
1 - norm.cdf(90, 80, 5)
```

```python
# 70'den az olma olasılığı
1 - norm.cdf(70, 80, 5)
```

```python
# 73'ten az olma olasılığı
norm.cdf(73, 80, 5)
```

```python
# 85 ile 90 arasında olma olasılığı
norm.cdf(90, 80, 5) - norm.cdf(85, 80, 5)
```
