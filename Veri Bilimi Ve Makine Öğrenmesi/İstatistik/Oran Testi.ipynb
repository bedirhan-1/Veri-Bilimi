# Tek Örneklem Oran Testi

# Problem 
#### Bir yazılım ile mecrada reklam verilmiş ve bu reklama ilişkin yazılım tarafından 0.125 dönüşüm oranı elde edildiği ifade edilmiştir. Fakat bu durum kontrol edilmek isteniyor. Çünkü bu yüksek bir oran ve gelirler incelendiğinde örtüşmüyor.

# Detaylar
* 500 kişi dış mecrada reklamalara tıklamış. 40 tanesi sitemize gelip alışveriş yapmış. 
* Örnek üzerinden ekde ediken dönüşüm oranı: 40/500 = 0.08


##### H0: p = 0.125
##### H1: p != 0.125

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
count = 40
nobs = 500
value = 0.125
```

```python
proportions_ztest(count, nobs, value)
```
