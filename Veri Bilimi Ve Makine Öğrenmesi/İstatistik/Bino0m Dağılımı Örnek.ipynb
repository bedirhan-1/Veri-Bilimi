# İş uygulaması : Reklam Harcama Optimizasyonu

#### Problem : 
Çeşitli mecralara reklam veriliyor, reklamların
tıklanma ve geri dönüşüm oranları optimize edilmeye çalışılıyor.
Buna yönelik belirli bir mecrada çeşitli senaryolara göre reklama 
tıklama olasılıkları hesaplanmak isteniliyor.

#### Detaylar : 
* bir mecrada reklam verilecek 
* Dağılım ve reklama tıklama olasılığı biliniyor (0.01)
* Soru : Reklamı 100 kişi gördüğünde 1, 5, 10 tıklanma olasılığı nedir?

#### Olasılıkların Hesaplanması Sonuçları:

* 1 kişi için = 0.37
* 5 kişi için = 0.00289779
* 10 kişi için = 0.00000007

# Program Yazarak Çözmek

#### Binom Dağılımı

```python
from scipy.stats import binom
```

```python
p = 0.01
n = 100
rv = binom(n, p)
print(rv.pmf(1))
print(rv.pmf(5))
print(rv.pmf(10))
```
