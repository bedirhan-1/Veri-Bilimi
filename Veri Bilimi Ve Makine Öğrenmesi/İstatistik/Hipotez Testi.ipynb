# Hipotez
Bir inanışı (sav, tahmin, vs) test etmek için kullanılan istatistiksel bir tekniktir

----------------------------------------------

# p-value Değeri
Hipotez testlerinin sonuçlarını değerlendirmek üzere programlar tarafından p-value değeri verilir. Bu değer üzerinden kolayca yorum yapabiliriz.

p < 0.05

İlgili H0 hipotezi reddedilmiş demektir.

α = 0.05 (kabul edilebilir hata miktarı)

------------------------------------------
# Hipotez Testi Adımları

1. Hipotezlerin kurulması ve yönlerinin belirlenmesi
2. Anlamlılık düzeyinin ve tablo değerinin belirlenmesi
3. Test istatistiğinin belirlenmesi ve test istatistiğinin hesaplanması
4. Hesaplanan test istatistiği ile α değerine karşılık gelen tablo değerinin karşılaştırılması.

Test istatistiği (Zh) > Tablo Değeri(Zt) ise H0 Red

5. Yorum
